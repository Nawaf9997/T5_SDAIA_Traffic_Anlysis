## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5));
sns.countplot(x='TrafficIndexLive', data=df, hue='TrafficIndexLive');
plt.title('Distrubution of The traffic index');
plt.xlabel('Traffic index');
plt.ylabel('Count');
ax = plt.gca();

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right");

plt.show();
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Task (2) Dataset.csv')

plt.figure(figsize=(10, 5))
sns.histplot(df['TrafficIndexLive'], kde=True, bins=30, color='skyblue')
plt.title('The_Traffic_Index')
plt.xlabel('TrafficIndexLive')
plt.ylabel('City')
plt.show()
```
